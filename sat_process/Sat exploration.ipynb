```python
%matplotlib notebook
import datetime as dt  # Python standard library datetime  module
import numpy as np
import netCDF4  # http://code.google.com/p/netcdf4-python/
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, addcyclic, shiftgrid

import s2_py as s2
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, mapping

import os
import pickle
import wget
import subprocess


import matplotlib.animation as animation
```

```python
data_dir = '/home/scott/sat_pickles'
picklefiles = {}
for dirpath, dirnames, filenames in os.walk(data_dir):
    for filename in [f for f in filenames if f.endswith(".pkl")]:
        picklefiles[filename] = (os.path.join(dirpath, filename))
```

```python
sorted(picklefiles.keys())
```

```python
len(picklefiles.keys()) / 3
```

```python
# len(2014 ... 2018) * 36 = 180 
# 5 mos * 3 / mo = 15, + 1 for June , so 196 checks out...
```

```python
years = range(2014, 2020)
year = str(2014)
#for year in years:
fapars ={}
for filename, fullpath in picklefiles.items():
    if filename.startswith('FAPAR'):
        dateStart = filename.find(str(year))
        date = filename[dateStart:dateStart+8]
        #print(filename, date)

        with open(fullpath, 'rb') as f:
            fapar = pickle.load(f)
        fapars[date] = fapar

fig = plt.figure()
title_obj = plt.title(year) #get the title property handler
ims = []
for key in sorted(fapars.keys()):
    im = plt.imshow(fapars[key] , animated = True,)
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval = 50, blit = True,
                               repeat_delay = 1000)
#writer = animation.writers['ffmpeg'](fps=1)
#ani.save(year + '.mp4',writer=writer,dpi=dpi)
plt.show()
```
